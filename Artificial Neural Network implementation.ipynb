```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data.csv')
del data['Unnamed: 32']
```

```python
data.head()
```

```python
data.shape
```

```python
X = data.iloc[:, 2:].values
y = data.iloc[:, 1].values
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
y = labelencoder_X_1.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import models
from keras import layers
```

```python
classifier = models.Sequential()
```

```python
classifier.add(layers.Dense(16, activation='tanh' ,input_dim=30,use_bias=True,name = 'input'))
classifier.add(layers.Dropout(rate=0.1))
```

```python
classifier.add(layers.Dense(16, activation='tanh',use_bias=True,name = 'hidden1'))
classifier.add(layers.Dropout(rate=0.1))
```

```python
classifier.add(layers.Dense(1,  activation='sigmoid',use_bias=True,name = 'output'))
```

```python
classifier.summary()
```

```python
classifier.weights
```

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = classifier.fit(X_train, y_train, batch_size=100, epochs=100,validation_split = 0.02)
```

```python
plt.plot(history.history['accuracy'])
plt.show()
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
print("Our accuracy is {}%".format(((cm[0][0] + cm[1][1])/114)*100))
```
